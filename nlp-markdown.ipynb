text = nltk.Text(tokens) 를 이용해서 위에서 리스트로 만들어준 text를 nltk.text 형태로 만들어준다. 형태는 유사하나 nltk 함수들을 이용해서 다양한 값들을 가져올 수 있다.

## 이번엔 nltk 라이버러리에 포함되어 있는 모비딕을 가져온다.

```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
raw
```

#### FreqDist는 Frequency Distribution으로서 어떤 스트링이 자주 나왔는지 알아내는 함수다. fdist.most_common(5)를 이용해 가장 자주 나온 스트링을 순서대로 5개를 보여준다.

#### fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha()) -  for loop를 돌리는데 if를 이용해 조건을 정해준다. isalpha는 알파벳인 경우만 가져오는 것이다. 따라서 아까와 달리 스페이스나 특수문자는 제외된다. 또 ch.lower를 통해 그것들을 소문자로 바꿔준다.

### [char for (char, count) in fdist.most_common(5)] 는 찾아낸 빈도가 가장 높은 단어들 중에서 알파벳만 나열해주는 것이다. (count 빼고)

```python
fdist = nltk.FreqDist(raw)
print(fdist.most_common(5))

fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())
print(fdist.most_common(5))
[char for (char, count) in fdist.most_common(5)]
```

```python
print(fdist.most_common(1))
```

```python
fdist = nltk.FreqDist(ch.upper() for ch in raw if ch.isalpha())
print(fdist.most_common(1))
```

```python
fdist = nltk.FreqDist(ch.upper() for ch in raw)
print(fdist.most_common(5))
```

```python
print(fdist)
```
